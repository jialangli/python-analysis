```python
train_data['便利设施_num'] = train_data['便利设施'].str.split(',')
int_equipments = []
for item in train_data['便利设施_num']:
    num = len(item)
    int_equipments.append(num)
train_data['便利设施_num'] = int_equipments
test_data['便利设施_num'] = test_data['便利设施'].str.split(',')
int_equipments_test = []
for item in test_data['便利设施_num']:
    num = len(item)
    int_equipments_test.append(num)
test_data['便利设施_num'] = int_equipments_test
```

```python
#转化为str
for key_words in['Internet','Kitchen','Pets','Air','Washer','Aid','Heating','Dryer','Essentials','Hangers','TV','parking','24_hour']:
    train_data[key_words]=train_data['便利设施'].str.contains(key_words).astype('str')
    test_data[key_words]=test_data['便利设施'].str.contains(key_words).astype('str')
```

```python
# 使用-999来处理民宿缺失值
```

```python
train_data['民宿周边'] = train_data['民宿周边'].fillna(-999)
test_data['民宿周边'] = test_data['民宿周边'].fillna(-999)
surround = []
for value in train_data['民宿周边']:
    if value == -999:
        new_value = 0
    else:
        new_value = 1
    surround.append(new_value)
train_data['民宿周边'] = surround

surround_test = []
for value in train_data['民宿周边']:
    if value == -999:
        new_value = 0
    else:
        new_value = 1
    surround_test.append(new_value)
train_data['民宿周边'] = surround_test
```

```python
# 关于房主是否有个人资料图片和房主身份是否验证使用0和1表示
```

```python
train_data['房主是否有个人资料图片']=train_data['房主是否有个人资料图片'].fillna('f')
train_data['房主身份是否验证']=train_data['房主身份是否验证'].fillna('f')
test_data['房主是否有个人资料图片']=test_data['房主是否有个人资料图片'].fillna('f')
test_data['房主身份是否验证']=test_data['房主身份是否验证'].fillna('f')
str_cols=['Internet','Kitchen','房主是否有个人资料图片','房主身份是否验证','Air','Pets','Washer','Aid','Heating','Dryer','Essentials','Hangers','TV','parking','24_hour']
le=LabelEncoder()
for col in str_cols:
    train_data[col]=le.fit_transform(train_data[col])
    test_data[col]=le.fit_transform(test_data[col])
```

```python
train_data['房主回复率']=train_data['房主回复率'].fillna('20%')
train_data['房主回复率']=train_data['房主回复率'].str[:-1]
test_data['房主回复率']=test_data['房主回复率'].fillna('20%')
test_data['房主回复率']=test_data['房主回复率'].str[:-1]
```

```python
# 将日期时间使用pd.to_datatime()方法来转化成日期时间数据类型，对于首次评价日期和最近评价日期和何时成为房主使用将缺失日期视为一个特殊的类别或者标记值，表示信息的缺失。比如2019-1-1
```

```python
train_data['首次评论日期']=train_data['首次评论日期'].fillna('2019-1-1')
train_data['最近评论日期']=train_data['最近评论日期'].fillna('2019-1-1')
test_data['首次评论日期']=test_data['首次评论日期'].fillna('2019-1-1')
test_data['最近评论日期']=test_data['最近评论日期'].fillna('2019-1-1')
time_cols=['首次评论日期','最近评论日期','何时成为房主']
for col in time_cols:
    train_data[col]=pd.to_datetime(train_data[col])
    test_data[col]=pd.to_datetime(test_data[col])
```

```python
# 经过相关资料搜索，邮政编码是一种用于识别特定地区或邮局的数字代码系统。但我们已经有城市，经纬度和民宿周边了，可以推出地区所在，邮编对于民宿价格预测没有多大影响，所以我决定把邮编这一列给删掉。
```

```python
# 删除名为"邮编"的列
train_data.drop('邮编', axis=1, inplace=True)
test_data.drop('邮编', axis=1, inplace=True)
```

```python
#处理数据缺失值和数据类型转换
cols=train_data.columns
cont_cols=['容纳人数','床的数量','卧室数量','洗手间数量','维度','经度','评论个数','民宿评分','房主回复率']
cate_cols=['Internet','Kitchen','Air','Pets','Washer','Aid','Heating','Dryer','Essentials','Hangers',"TV",'parking','24_hour','床的类型','民宿周边','取消条款','所在城市','清洁费','是否支持随即预订','房产类型','房型','房主是否有个人资料图片', '房主身份是否验证']
features=cont_cols+cate_cols
train_data[cate_cols]=train_data[cate_cols].astype('category')
test_data[cate_cols]=test_data[cate_cols].astype('category')
for col in cate_cols:
    train_data[col]=train_data[col].fillna(train_data[col].mode())
    test_data[col]=test_data[col].fillna(test_data[col].mode())
    
for col in ['洗手间数量','床的数量','卧室数量','民宿评分']:
    train_data[col]=train_data[col].fillna(train_data[col].median())
    test_data[col]=test_data[col].fillna(test_data[col].median())
train_data=train_data.dropna().reset_index()
```

```python
#数据预处理和特征缩放
X_train=train_data[features]
scaler=RobustScaler()
scaler.fit(X_train[cont_cols])
X_train[cont_cols]=scaler.transform(X_train[cont_cols])
X_test=test_data[features]
X_test[cont_cols]=scaler.transform(X_test[cont_cols])
#转换为数字类型
y_train=train_data['价格'].astype('int')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_validation, y_train, y_validation = train_test_split(X_train, y_train, random_state=888, test_size=0.2)
```

```python
from catboost import CatBoostRegressor, Pool
train_pool = Pool(X_train, y_train)
validation_pool = Pool(X_validation, y_validation)
model = CatBoostRegressor()
model.fit(train_pool, eval_set=validation_pool,verbose=False)
y_pred = model.predict(X_validation)
mse = mean_squared_error(y_validation, y_pred)
print("均方误差（MSE）：", mse)
rmse = np.sqrt(mse)
print("均方根误差（RMSE）：", rmse)
mae = mean_absolute_error(y_validation, y_pred)
print("平均绝对误差（MAE）：", mae)
r2 = r2_score(y_validation, y_pred)
print("决定系数（R^2）：", r2)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_validation)
mse = mean_squared_error(y_validation, y_pred)
print("均方误差（MSE）：", mse)
rmse = np.sqrt(mse)
print("均方根误差（RMSE）：", rmse)
mae = mean_absolute_error(y_validation, y_pred)
print("平均绝对误差（MAE）：", mae)
r2 = r2_score(y_validation, y_pred)
print("决定系数（R^2）：", r2)
```

```python
# 创建决策树回归模型
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_validation)
mse = mean_squared_error(y_validation, y_pred)
print("均方误差：", mse)
rmse = np.sqrt(mse)
print("均方根误差（RMSE）：", rmse)
mae = mean_absolute_error(y_validation, y_pred)
print("平均绝对误差（MAE）：", mae)
r2 = r2_score(y_validation, y_pred)
print("决定系数（R2）：", r2)
```

```python
# 创建随机森林回归模型
regressor = RandomForestRegressor()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_validation)
mse = mean_squared_error(y_validation, y_pred)
print("均方误差：", mse)
rmse = np.sqrt(mse)
print("均方根误差（RMSE）：", rmse)
mae = mean_absolute_error(y_validation, y_pred)
print("平均绝对误差（MAE）：", mae)
r2 = r2_score(y_validation, y_pred)
print("决定系数（R2）：", r2)
```
